```python
import numpy as np
import pandas as pd

from pandas import Series, DataFrame
```

```python
# how about data frame
# not this is cool gets data from clipboard
import webbrowser
website = 'http://en.wikipedia.org/wiki/NFL_win-loss_records'
webbrowser.open(website)
```

```python
#copy the columns you want from the table in the page then

nfl_frame = pd.read_clipboard() #read clipboard
nfl_frame
```

```python
# We can grab the oclumn names with .columns
nfl_frame.columns
```

```python
#Lets see some specific data columns
DataFrame(nfl_frame,columns=['Team','Won','Lost'])
```

```python
#What happens if we ask for a column that doesn't exist?
DataFrame(nfl_frame,columns=['Team','Won','Lost','Stadium'])
```

```python
#We can retrieve individual columns
nfl_frame.Team
```

```python
#We can retrieve rows through indexing
nfl_frame.loc[3]
```

```python
#We can also assign value sto entire columns
nfl_frame['Stadium']="Levi's Stadium" #Careful with the ' here
nfl_frame
```

```python
#Putting numbers for stadiums
nfl_frame["Stadium"] = np.arange(5)

#Show
nfl_frame
```

```python
#Adding a Series to a DataFrame
stadiums = Series(["Levi's Stadium","AT&T Stadium"],index=[4,1])

#Now input into the nfl DataFrame
nfl_frame['Stadium']=stadiums

#Show
nfl_frame
```

```python
#We can also delete columns
del nfl_frame['Stadium']

nfl_frame
```

```python
# artithmetic DataFrames!
dframe1 = DataFrame(np.arange(4).reshape(2,2),columns=list('AB'),index=['NYC','LA'])

#Show
dframe1
```

```python
#Second DataFrame
dframe2 = DataFrame(np.arange(9).reshape(3,3),columns=list('ADC'),index=['NYC','SF','LA'])

#Show
dframe2
```

```python
#What happens when we add them together?

dframe1 + dframe2
```

```python
#What if we want to replace the NaN values
# Then we can use .add()

dframe1.add(dframe2,fill_value=0)
```
