```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
#read data we want to apply some agg
#a csv on white wine

df_wines = pd.read_csv('winequality-white.csv', sep=';')

#check some data
df_wines.head()
```

```python
#create a function to agg
def max_to_min(arr):
    return arr.max() - arr.min()


#use agg ang emplement fuction
df_wines.agg(max_to_min)
```

```python
#if you group by quality then agg you get
wwines = df_wines.groupby('quality')

wwines.agg(max_to_min)
```

```python
wwines.agg('mean')
```
